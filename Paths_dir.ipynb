```python
def xfind_md_files(start_path):
    """Recursively find all .md files starting from the given path."""
    md_files = []
    for root, dirs, files in os.walk(start_path):
        for file in files:
            if file.endswith(".md"):
                full_path = os.path.join(root, file)
                md_files.append(full_path)
    return md_files
```

```python
import os

def find_md_files_with_class_in_name(start_path):
    """Recursively find .md files whose filenames contain the word 'class'."""
    md_files = []
    for root, dirs, files in os.walk(start_path):
        for file in files:
            if file.endswith(".md") and "LIC" in file.lower():
                full_path = os.path.join(root, file)
                md_files.append(full_path)
    return md_files
```
