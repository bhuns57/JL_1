### Inheritance-Based Classification Scaffold
**ðŸ”§ Base Class: CompoundBase**

```python
class CompoundBase:
    def __init__(self, gsn, name, dose=None, flags=None):
        self.gsn = gsn
        self.name = name
        self.dose = dose
        self.flags = flags or []

    def describe(self):
        return f"{self.name} (GSN: {self.gsn}) â€” Dose: {self.dose} mg"

# Generation subclasses
class Generation1(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 1

class Generation2(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 2

class Generation3(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 3

class Generation4(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 4

class Generation5(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 5
```

### ðŸŒ¿ Subclass Example: Generation1, Generation2, etc.

```python
class Generation1(CompoundBase):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 1

class Generation2(Generation1):
    def __init__(self, gsn, name, dose=None, flags=None):
        super().__init__(gsn, name, dose, flags)
        self.level = 2
```

### ðŸ§ª Instantiate a Full Lineage

```python
g5 = Generation5("05", "CoQ10", dose=100)
g4 = Generation4("04", "MitoSupport")
g3 = Generation3("03", "Energy")
g2 = Generation2("02", "Metabolic")
g1 = Generation1("01", "Health")

lineage = [g1, g2, g3, g4, g5]
id10 = ''.join([node.gsn for node in lineage])
print("ID10:", id10)
```
