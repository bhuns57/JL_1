# example data set

```python
import pandas as pd

# Sample data
data = {
    'date': pd.date_range(start='2025-06-01', periods=60),
    'phase_angle': [5.8, 5.7, 5.6, 5.4, 5.2, 5.1, 5.0, 5.2, 5.4, 5.6] * 6,
    'ICW': [22.0, 21.8, 21.6, 21.4, 21.2, 21.0, 20.8, 21.0, 21.2, 21.4] * 6,
    'ECW': [15.0, 15.2, 15.4, 15.6, 15.8, 16.0, 16.2, 16.0, 15.8, 15.6] * 6,
    'right_leg_mass': [8.0, 7.9, 7.8, 7.7, 7.6, 7.5, 7.4, 7.5, 7.6, 7.7] * 6,
    'trunk_mass': [20.0, 19.8, 19.6, 19.4, 19.2, 19.0, 18.8, 19.0, 19.2, 19.4] * 6,
}
df = pd.DataFrame(data)

# Antibiotic log
antibiotics = [
    {'name': 'Amox-Clav', 'start': '2025-06-10', 'end': '2025-06-17'},
    {'name': 'TMP-SMX', 'start': '2025-07-01', 'end': '2025-07-07'}
]
df
```

# 1. Trend-Based Mitochondrial Stress Detection

```python
df['phase_angle_rolling'] = df['phase_angle'].rolling(window=7).mean()
df['ICW_ECW_ratio'] = df['ICW'] / df['ECW']
df['ICW_ECW_rolling'] = df['ICW_ECW_ratio'].rolling(window=7).mean()

# Flagging drops
df['phase_decline'] = df['phase_angle_rolling'].diff()
df['ratio_decline'] = df['ICW_ECW_rolling'].diff()

#df
```

# 📊 2. Antibiotic Impact Modeling

```python
df['antibiotic'] = None
for abx in antibiotics:
    mask = (df['date'] >= pd.to_datetime(abx['start'])) & (df['date'] <= pd.to_datetime(abx['end']))
    df.loc[mask, 'antibiotic'] = abx['name']
    
# df.loc[mask, 'antibiotic']
```

# 🧠 3. Treatment Optimization

```python
summary = df.groupby('antibiotic').agg({
    'phase_decline': 'min',
    'ICW_ECW_ratio': ['min', 'max'],
    'right_leg_mass': 'mean',
    'trunk_mass': 'mean'
})

summary
```

# 🧰 Optional: Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['phase_angle_rolling'], label='Phase Angle (7-day avg)')
plt.plot(df['date'], df['ICW_ECW_rolling'], label='ICW/ECW Ratio (7-day avg)', linestyle='--')

for abx in antibiotics:
    plt.axvspan(pd.to_datetime(abx['start']), pd.to_datetime(abx['end']), color='red', alpha=0.2, label=abx['name'])

plt.legend()
plt.title('Mitochondrial Stress and Antibiotic Overlay')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

# Diagnostic dashboard for mitochondrial stress and antibiotics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set seaborn style
sns.set(style="whitegrid")

# Create output directory if it doesn't exist
output_dir = "JL_1"
os.makedirs(output_dir, exist_ok=True)

# Simulated dataset structure
np.random.seed(42)
dates = pd.date_range(start="2025-06-01", end="2025-08-10", freq="D")
data = pd.DataFrame({
    "Date": dates,
    "Mitochondrial_Stress": np.random.normal(loc=50, scale=10, size=len(dates)),
    "Bioenergetic_Fingerprint": np.random.normal(loc=100, scale=15, size=len(dates)),
    "Urine_pH": np.random.normal(loc=6.5, scale=0.3, size=len(dates)),
    "Symptoms": np.random.choice(["None", "Fatigue", "Pain", "Brain Fog"], size=len(dates))
})

# Simulated antibiotic log
antibiotic_periods = [
    ("2025-06-10", "2025-06-20"),
    ("2025-07-05", "2025-07-15"),
    ("2025-08-01", "2025-08-07")
]

# Create dashboard plot
fig, axs = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

# Plot mitochondrial stress
sns.lineplot(data=data, x="Date", y="Mitochondrial_Stress", ax=axs[0], color="red")
axs[0].set_title("Mitochondrial Stress Over Time")
for start, end in antibiotic_periods:
    axs[0].axvspan(pd.to_datetime(start), pd.to_datetime(end), color='gray', alpha=0.3)

# Plot bioenergetic fingerprint
sns.lineplot(data=data, x="Date", y="Bioenergetic_Fingerprint", ax=axs[1], color="blue")
axs[1].set_title("Bioenergetic Fingerprint Over Time")
for start, end in antibiotic_periods:
    axs[1].axvspan(pd.to_datetime(start), pd.to_datetime(end), color='gray', alpha=0.3)

# Plot urine pH
sns.lineplot(data=data, x="Date", y="Urine_pH", ax=axs[2], color="green")
axs[2].set_title("Urine pH Over Time")
for start, end in antibiotic_periods:
    axs[2].axvspan(pd.to_datetime(start), pd.to_datetime(end), color='gray', alpha=0.3)

# Plot symptoms timeline
symptom_colors = {"None": "lightgray", "Fatigue": "orange", "Pain": "red", "Brain Fog": "purple"}
symptom_series = data.set_index("Date")["Symptoms"]
axs[3].set_title("Symptom Timeline")
for date, symptom in symptom_series.items():
    axs[3].axvline(date, color=symptom_colors[symptom], alpha=0.6)
axs[3].set_ylabel("Symptoms")
axs[3].set_yticks([])

# Final layout adjustments
plt.tight_layout()
import os
# Save to a local folder
output_dir = os.path.expanduser("~/diagnostic_outputs")
os.makedirs(output_dir, exist_ok=True)

# Save the figure
plt.savefig(os.path.join(output_dir, "diagnostic_dashboard.png"))





# plt.savefig("/mnt/data/diagnostic_dashboard.png")
plt.close()

print("Dashboard saved to /mnt/data/diagnostic_dashboard.png")
```
