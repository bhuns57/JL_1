```python
import pandas as pd
import numpy as np

def load_csv_columns_as_arrays(filename: str, index_col=None) -> dict:
    df = pd.read_csv(filename, index_col=index_col)
    col_dict = {}
    for col in df.columns:
        col_dict[col] = df[col].to_numpy()
    return col_dict
```
